```python
import pandas as pd
import numpy as np

# Assuming X and df already exist as per your previous code
features = ['Ammonia_ppm', 'CO2_ppm', 'Oxygen_percent',
            'Temperature_C', 'Humidity_percent', 'Light_lux', 'Methane_ppm']

def detect_outliers_iqr(df, features):
    outlier_indices = set()  # to collect unique rows
    
    for col in features:
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1
        
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        
        col_outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)].index
        outlier_indices.update(col_outliers)
        
        print(f"{col:18} → Outliers: {len(col_outliers)} (bounds: {lower_bound:.2f} – {upper_bound:.2f})")
    
    return sorted(list(outlier_indices))

# Kureba outliers
outlier_rows = detect_outliers_iqr(df, features)

print(f"\nUmubare w'amakuru yose yagaragaye nk'outliers (ku feature icyo aricyo cyose): {len(outlier_rows)}")
print("Indices z'outliers:", outlier_rows)

# Reba ibimenyetso by'indwara kuri outliers
if outlier_rows:
    print("\nIndwara ziri kuri outliers:")
    print(df.loc[outlier_rows, 'Disease'].value_counts())
    print("\nUrugero rw'outliers (mbere 5):")
    print(df.loc[outlier_rows].head())

# Kuguramo outliers (remove rows flagged in any feature)
df_clean_iqr = df.drop(outlier_rows)
print(f"\nOriginal: {len(df)} rows → After IQR cleaning: {len(df_clean_iqr)} rows")
```

# Real-Time and Manual Prediction from MySQL Database
This section demonstrates how to use your trained model (`Chickhelf.pkl`) to make predictions using sensor data from your MySQL database, both for real-time (today's) and manual (date range) scenarios.

```python
import os
import pandas as pd
import mysql.connector
import joblib
from datetime import date

# Read database credentials from environment variables.
# Set MYSQL_HOST, MYSQL_USER, MYSQL_PASSWORD, and MYSQL_DATABASE before running these cells.
db_host = os.getenv("MYSQL_HOST", "localhost")
db_user = os.getenv("MYSQL_USER")
db_password = os.getenv("MYSQL_PASSWORD")
db_name = os.getenv("MYSQL_DATABASE")

missing_values = [
    name
    for name, value in [
        ("MYSQL_USER", db_user),
        ("MYSQL_PASSWORD", db_password),
        ("MYSQL_DATABASE", db_name),
    ]
    if not value
]

if missing_values:
    conn = None
    print(
        "MySQL cells are skipped until these environment variables are set: "
        + ", ".join(missing_values)
    )
    print("Set them in VS Code, then rerun this cell and the query cells below.")
else:
    conn = mysql.connector.connect(
        host=db_host,
        user=db_user,
        password=db_password,
        database=db_name,
    )
    print("MySQL connection established successfully.")
```

```python
# --- Real-Time Prediction: All Data from Today ---
# Adjust feature order to match your model: [nh3, co2, h2s, co, o2, temperature, humidity, light_lux, ch4]
# If 'co' (carbon monoxide) is missing, fill with 0

if conn is None:
    print("Skipping real-time MySQL prediction because the database connection is not configured.")
else:
    today = date.today().strftime('%Y-%m-%d')
    query_today = '''
    SELECT nh3, co2, h2s, 0 as co, o2, temperature, humidity, light_lux, ch4
    FROM sensors
    WHERE DATE(recorded_at) = %s
    '''

    df_today = pd.read_sql(query_today, conn, params=[today])

    # Load the trained model
    model = joblib.load('Chickhelf.pkl')

    if not df_today.empty:
        predictions_today = model.predict(df_today)
        df_today['Predicted_Disease'] = predictions_today
        print(df_today)
    else:
        print('No sensor data found for today.')
```

```python
# --- Manual Prediction: Data from a Selected Date Range ---
# User specifies start_date and end_date (YYYY-MM-DD)

if conn is None:
    print("Skipping manual MySQL prediction because the database connection is not configured.")
else:
    start_date = '2026-03-12'  # example, change as needed
    end_date = '2026-03-16'    # example, change as needed

    query_range = '''
    SELECT nh3, co2, h2s, 0 as co, o2, temperature, humidity, light_lux, ch4
    FROM sensors
    WHERE DATE(recorded_at) BETWEEN %s AND %s
    '''

    df_range = pd.read_sql(query_range, conn, params=[start_date, end_date])

    if not df_range.empty:
        predictions_range = model.predict(df_range)
        df_range['Predicted_Disease'] = predictions_range
        print(df_range)
    else:
        print(f'No sensor data found between {start_date} and {end_date}.')
```
